```python
df2 = df.drop(df.index[10])
```

```python
df2
```

```python
trace0 = go.Scatter(
    x = df2["Row_labels"],
    y = df2["Average_of_stars"],
    name = 'Reviews',
    mode = 'markers',
    marker = dict(
        size = df2["Count_of_review_count"] / 1000,
        color = 'red',
        line = dict(
            width = 3,
            color = ('red')
        )
    )
)

data = [trace0]

layout = dict(title = 'Yelp Reviews: Business Ratings',
             yaxis = dict(zeroline = False),
             xaxis = dict(zeroline = False)
            )
fig = dict(data=data, layout=layout)
py.iplot(fig, filename='Yelp_Reviews')
```
